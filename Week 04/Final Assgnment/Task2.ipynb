```python
def access_element(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    return arr[0]

def insert_element(arr):
    # Time Complexity: O(1) amortized
    # Space Complexity: O(1)
    arr.append(1)

def delete_element(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)
    del arr[0]

def update_element(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    arr[0] = 2

def traverse_array(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)
    for element in arr:
        pass

def sort_array(arr):
    # Time Complexity: O(n log n)
    # Space Complexity: O(1)
    arr.sort()

def search_array(arr, target):
    # Time Complexity: O(n)
    # Space Complexity: O(1)
    return target in arr
```

```python
def access_element_2d(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    return arr[0][0]

def insert_element_2d(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    arr.append([1, 2, 3])

def delete_element_2d(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)
    del arr[0]

def update_element_2d(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    arr[0][0] = 2

def traverse_array_2d(arr):
    # Time Complexity: O(n*m)
    # Space Complexity: O(1)
    for row in arr:
        for element in row:
            pass

def sort_array_2d(arr):
    # Time Complexity: O(n*m log (n*m))
    # Space Complexity: O(1)
    for row in arr:
        row.sort()

def search_array_2d(arr, target):
    # Time Complexity: O(n*m)
    # Space Complexity: O(1)
    for row in arr:
        if target in row:
            return True
    return False
```

```python
def access_element_3d(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    return arr[0][0][0]

def insert_element_3d(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    arr.append([[1, 2], [3, 4]])

def delete_element_3d(arr):
    # Time Complexity: O(n)
    # Space Complexity: O(1)
    del arr[0]

def update_element_3d(arr):
    # Time Complexity: O(1)
    # Space Complexity: O(1)
    arr[0][0][0] = 2

def traverse_array_3d(arr):
    # Time Complexity: O(n*m*p)
    # Space Complexity: O(1)
    for layer in arr:
        for row in layer:
            for element in row:
                pass

def sort_array_3d(arr):
    # Time Complexity: O(n*m*p log (n*m*p))
    # Space Complexity: O(1)
    for layer in arr:
        for row in layer:
            row.sort()

def search_array_3d(arr, target):
    # Time Complexity: O(n*m*p)
    # Space Complexity: O(1)
    for layer in arr:
        for row in layer:
            if target in row:
                return True
    return False
```
