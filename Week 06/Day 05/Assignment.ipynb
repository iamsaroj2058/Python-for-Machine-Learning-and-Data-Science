- <li> let's plot graphs for four different equations using independent variables ranging from 100 to 200. Plot the graphs for Simple Linear Regression, Multiple Linear Regression, Simple Polynomial Regression (Degree 2), and Multiple Polynomial Regression (Degree 3)</li>

- <li> Additionally, let's figure out how to obtain continuous values from the attached CSV files using you'll</li>

- <li>For Position_Salaries.csv, you'll perform Simple Polynomial Regression (Degree 2) and Multiple Polynomial Regression (Degree 3).</li>
- <li>For Salary_Data.csv, you'll conduct Simple Linear Regression.</li>
- <li>For 50_Startups.csv, you'll employ Multiple Linear Regression.</li>

```python
#Question no.01

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate independent variables ranging from 100 to 200
X = np.arange(100, 201).reshape(-1, 1)

# Generate corresponding dependent variables for Simple Linear Regression
y_simple_linear = 0.5 * X + 30

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X, y_simple_linear, color='blue', label='Simple Linear Regression')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question no.2
# Generate corresponding dependent variables for Multiple Linear Regression
y_multi_linear = 0.3 * X + 0.2 * X**2 - 5000 + np.random.normal(0, 500, size=X.shape)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X, y_multi_linear, color='green', label='Multiple Linear Regression')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Multiple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Question number 03

from sklearn.preprocessing import PolynomialFeatures

# Generate corresponding dependent variables for Simple Polynomial Regression (Degree 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly, y_simple_linear)
y_simple_poly = lin_reg_poly.predict(X_poly)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X, y_simple_linear, color='blue', label='Simple Linear Regression')
plt.plot(X, y_simple_poly, color='red', label='Simple Polynomial Regression (Degree 2)')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Simple Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question number
# Generate corresponding dependent variables for Multiple Polynomial Regression (Degree 3)
poly_features_multi = PolynomialFeatures(degree=3)
X_poly_multi = poly_features_multi.fit_transform(X)
lin_reg_multi_poly = LinearRegression()
lin_reg_multi_poly.fit(X_poly_multi, y_multi_linear)
y_multi_poly = lin_reg_multi_poly.predict(X_poly_multi)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X, y_multi_linear, color='green', label='Multiple Linear Regression')
plt.plot(X, y_multi_poly, color='orange', label='Multiple Polynomial Regression (Degree 3)')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Multiple Polynomial Regression (Degree 3)')
plt.legend()
plt.grid(True)
plt.show()
```
