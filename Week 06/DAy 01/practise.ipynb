```python
import matplotlib.pyplot as plt

x = ["python","C","C++","Java"]
y = [100,120,150,180]
# for label and font size
plt.xlabel("language",fontsize=10)

# for y axis label and font size
plt.ylabel("Number Of People",fontsize=10)

# fon  title
plt.title("No of Programmers")

# for make it colourfull
c=["g","r","y","b"]
# for plotting the bar chart
plt.bar(x,y,color=c)
plt.show()
```
